```python
l1 = 1
l2 = 2
l3 = 3


v1 = np.array([4, 5, 1])
v2 = np.array([-4, 0, -4])
v3 = np.array([1, 3, 2])

l1 * v1 + l2 * v2 + l3 * v3
```

```python
ll1 = [l1, l2, l3]
ll2 = [v1, v2, v3]

res = 0

for i in range(0, 3, 1) :
    res += ll1[i] * ll2[i]

res
```

```python
a = np.array([3, 4])
b = np.array([1, 2])

def getPara(t, r) :
    return r * (t.dot(r) / r.dot(r))

def getOrth(r, para) :
    return r - para

print(b.dot(getOrth(b, getPara(a, b))))
```

```python
import matplotlib.pyplot as plt

A = np.array([2, 1])
B = np.array([1, 0.5])

xlim = [-2, 2]
ylim = [-1, 1]

scalars = np.random.uniform(low = -1, high = 1, size = 100)

plt.figure(figsize=(4,2))

for sc in scalars :
    p = A*sc
    q = B*sc
    plt.plot(p[0], p[1], 'o', color = "red", markersize = 3)
    plt.plot(q[0], q[1], 'o', color = "blue", markersize = 3)

plt.xlim(xlim)
plt.ylim(ylim)
plt.grid()
plt.show()
```

```python
A = np.array([1, 3])
B = np.array([2, 7])

xlim = [-20, 20]
ylim = [-10, 10]

plt.figure(figsize=(40,20))

for i in range(0, 2000000, 1) :
    p = A * np.random.uniform(low = -10, high = 10, size = 1) \
        + B * np.random.uniform(low = -10, high = 10, size = 1)
    plt.plot(p[0], p[1], 'o', color = "red", markersize = 10)

plt.xlim(xlim)
plt.ylim(ylim)

plt.grid()
plt.show()
```
