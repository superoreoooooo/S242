```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([[1, 4], [2, 8]])
Ainv = np.linalg.inv(A)

print(A @ Ainv)
```

```python
def matrixMinor(arr, i, j) :
    return np.delete(np.delete(arr, i - 1, axis=0), j - 1, axis=1)

A = np.arange(1, 10).reshape(3, 3)

print(A)
A11 = matrixMinor(A, 1, 1)

print(A11)
```

```python
T = np.random.randint(-10, 11, size=(40, 4))
TtT = T.T @ T
TtTinv = np.linalg.inv(TtT)
print(np.allclose(TtTinv @ TtT, np.eye(4)))
```

```python
import matplotlib.cm as cm

T = np.random.randint(-10, 11, size=(40, 4))
TtT = T.T @ T
TtTinv = np.linalg.inv(TtT)

LT = TtTinv @ T.T @ T

plt.imshow(LT, cmap=cm.cool)
plt.show()
```

```python
A = np.array([[1, 4], [2, 8]])

A_pinv = np.linalg.pinv(A)

print(A_pinv)

print(np.allclose(A, A@A_pinv@A))
print(np.allclose(A_pinv, A_pinv@A@A_pinv))
```

```python
T = np.array([[1,.5],[0,.5]]) # Transformation matrix
# define the set of points (a circle)
theta = np.linspace(0,2*np.pi-2*np.pi/20,20)
origPoints = np.vstack( (np.cos(theta),np.sin(theta)) )
# apply transformation
transformedPoints = T @ origPoints
backTransformed = np.linalg.inv(T) @ transformedPoints # T @ T_inv @ origPoints => I @ origPoints => origPoints
# plot the points
plt.figure(figsize=(6,6))
plt.plot(origPoints[0,:],origPoints[1,:],'ko',label='Original')
plt.plot(transformedPoints[0,:],transformedPoints[1,:],'s', color=[.1, .1, .8],label='Transformed')
plt.plot(backTransformed[0,:],backTransformed[1,:],'rx',markersize=15, color=[.6,.1,.1],label='Inverse-transformed')
plt.axis('square')
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.legend()
plt.savefig('Figure_07_06.png',dpi=300)
plt.show()
```
