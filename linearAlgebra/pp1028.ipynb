```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
scailingVars = np.linspace(0, 50, 40)

nExp = 10

matrixNorms = np.zeros((len(scailingVars), nExp))

for si in range(len(scailingVars)) :
    for expi in range(nExp) :
        R = np.random.randn(10, 10) * scailingVars[si]

        matrixNorms[si, expi] = np.linalg.norm(R, "fro")

plt.plot(scailingVars, np.mean(matrixNorms, axis = 1), 'ko-')
plt.xlabel('Matrix Scalar')
plt.ylabel('Matrix Frobenius norm')
plt.show()

print(matrixNorms[0, :])
```

```python
scailingVars = np.linspace(0, 50, 40)

nExp = 1000

matrixNorms = np.zeros((len(scailingVars), nExp))

for si in range(len(scailingVars)) :
    for expi in range(nExp) :
        R = np.random.randn(10, 10) * scailingVars[si]

        matrixNorms[si, expi] = np.linalg.norm(R, "fro")


plt.plot(scailingVars, matrixNorms, 'ko-')
plt.plot(scailingVars, np.mean(matrixNorms, axis = 1), 'ko-', color = 'red')
plt.xlabel('Matrix Scalar')
plt.ylabel('Matrix Frobenius norm')
plt.show()

print(matrixNorms[0, :])
```

```python
lst = np.random.randn(3, 3)

fro = np.linalg.norm(lst)

trc = np.sqrt(np.trace(lst.T @ lst))

print(fro, trc, fro == trc)
```
