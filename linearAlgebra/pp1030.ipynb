```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
import scipy.linalg


A = np.array([[1, -1], [-2, 2]])
B = np.array([[1, -1], [-2, 3]])

print(scipy.linalg.null_space(A))
print(scipy.linalg.null_space(B))
```

```python
scalars = np.random.uniform(low=-1, high=1, size=(40, 2))
#print(scalars)

A = np.array([[1, -1], [-2, 2]])
plt.figure(figsize=(4, 4))

for s in scalars :
    p = A * s
    pp = A.T * s
    plt.plot(p[0], p[1], 'o', color='red', markersize=3)
    plt.plot(pp[0], pp[1], 'o', color='blue', markersize=3)

plt.plot([0, scipy.linalg.null_space(A)[0, 0]], [0, scipy.linalg.null_space(A)[1, 0]], color='red', markersize=3)
plt.plot([0, -scipy.linalg.null_space(A.T)[0, 0]], [0, -scipy.linalg.null_space(A.T)[1, 0]], color='blue', markersize=3)

plt.grid()
plt.show()
```
